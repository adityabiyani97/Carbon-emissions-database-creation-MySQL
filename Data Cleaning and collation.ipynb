```python
import numpy as np
import pandas as pd
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
table1 = pd.read_csv('energy-related-co2.csv',  engine = 'python', skipfooter = 1)
table1
```

```python
# extract the state names and repeat them for the years 1990 - 2017 thus, repitition = 28.
sname = np.array(np.repeat(table1['State'], 28))

# repeat the array 1990-2017 51 time for all the unique states. 
year = np.concatenate([np.array(table1.columns[1:].astype(int))]*51)
```

```python
sname, year
```

```python
# extract all the rows for every state flatten them out. 
x1 = np.array(table1.iloc[:, 1:])
energy_related_co2 = np.concatenate(x1)
```

```python
table2 = pd.read_csv('percapita-energy-related-co2.csv')
table2
```

```python
x2 = np.array(table2.iloc[:, 1:])
percapita_energy_related_co2 = np.concatenate(x2)
```

```python
table3 = pd.read_csv('energy-intensity.csv')
table3
```

```python
x3 = np.array(table3.iloc[:, 1:])
energy_intensity = np.concatenate(x3)
```

```python
table4 = pd.read_csv('carbon-intensity-energy-supply.csv')
table4
```

```python
x4 = np.array(table4.iloc[:, 1:])
carbon_intensity_energy_supply = np.concatenate(x4)
```

```python
'''Note: If there are commas in your numbering system, the python will convert those numbers into strings. 
   So edit them using format option in CSV/ Excel file itself'''

table5 = pd.read_csv('carbon-intensity-economy.csv')
table5
```

```python
x5 = np.array(table5.iloc[:, 1:])
carbon_intensity_economy = np.concatenate(x5)
```

```python
data_columns = np.array([year, energy_related_co2, percapita_energy_related_co2, 
                         energy_intensity, carbon_intensity_energy_supply, carbon_intensity_economy, sname]).T
data_columns
```

```python
carbon_emission = pd.DataFrame(data_columns, columns = ['year', 'energy_related_co2', 'percapita_energy_related_co2', 
                         'energy_intensity', 'carbon_intensity_energy_supply', 'carbon_intensity_economy', 'Sname'])
```

```python
carbon_emission.to_csv('carbon-emissions.csv', index = False)
```
